```python
import matplotlib.pyplot as plt
import pandas as pd
data = dataset = pd.read_csv('/content/drivers.csv')
# Creating a DataFrame
df = pd.DataFrame(data)
# Displaying the DataFrame
print(df)
print(df.columns)
drivers = data['Driver']
Pos = data['Pos']

# Creating the bar plot
plt.bar(drivers, Pos)

# Add labels and title
plt.xlabel('Driver')
plt.ylabel('Pos')
plt.title('Bar Plot')

# Show the plot
plt.show()
```
